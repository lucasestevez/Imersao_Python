```python
import pandas as pd
import plotly.express as px
```

```python
pd.options.display.float_format = "{:.2f}".format
```

```python
df_principal = pd.read_excel("/content/Aula 3 - Imersão Python.xlsx", sheet_name="Principal")
df_principal
```

```python
df_total_acoes = pd.read_excel("/content/Aula 3 - Imersão Python.xlsx", sheet_name="Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("/content/Aula 3 - Imersão Python.xlsx", sheet_name="Ticker")
df_ticker
```

```python
df_chatGPT = pd.read_excel("/content/Aula 3 - Imersão Python.xlsx", sheet_name="ChatGPT")
df_chatGPT
```

```python
df_analises = pd.read_excel("/content/Aula 3 - Imersão Python.xlsx", sheet_name="Análises")
df_analises
```

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={"Último (R$)":"valor_final", "Var. Dia (%)": "var_dia_pct"}).copy()
df_principal
```

```python
df_principal['var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['var_inicial'] = df_principal['valor_final'] / (df_principal['var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on="Ativo", right_on="Código", how="left")
df_principal
```

```python
df_principal = df_principal.drop(columns=["Código"])
df_principal
```

```python
df_principal["var_acoes"] = (df_principal["valor_final"] - df_principal["var_inicial"]) * df_principal["Qtde. Teórica"]
df_principal
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={"Qtde. Teórica":"qtd_teorica"}).copy()
df_principal
```

```python
df_principal['resultado'] = df_principal['var_acoes'].apply(lambda x: 'Subiu' if x > 0 else('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on="Ativo", right_on="Ticker", how="left")
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatGPT, left_on="Nome", right_on="Nome da Empresa", how="left")
df_principal = df_principal.drop(columns=['Nome da Empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={"Idade (em anos)":"Ano_de_fundacao"}).copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={"Ano_de_fundacao":"Idade"}).copy()
df_principal
```

```python
df_principal['Cat_idade'] = df_principal['Idade'].apply(lambda x: 'Mais de 100' if x > 100 else('Menos de 50' if x < 50 else 'Entre 50 e 100'))
df_principal
```

```python
# Calcular o valor máximo, mínimo e médio da coluna 'var_acoes'
maximo = df_principal['var_acoes'].max()
minimo = df_principal['var_acoes'].min()
media_total = df_principal['var_acoes'].mean()

# Calcular a média da coluna 'var_acoes' para os casos em que o resultado é 'Subiu' e 'Desceu'
media_subiu = df_principal[df_principal['resultado'] == 'Subiu']['var_acoes'].mean()
media_desceu = df_principal[df_principal['resultado'] == 'Desceu']['var_acoes'].mean()

# Exibir os resultados
print(f"Maior\tR$ {maximo:,.2f}")
print(f"Menor\tR$ {minimo:,.2f}" )
print(f"Média\tR$ {media_total:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby("Segmento")['var_acoes'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby("resultado")['var_acoes'].sum().reset_index()
df_analise_saldo
```

```python
fig_analise_saldo = px.bar(df_analise_saldo, x='resultado', y='var_acoes',
                           text='var_acoes', title="Variação R$ por Resultado")
fig_analise_saldo.update_traces(marker_color='#a3b18a', marker_line_color='#283618',
                  marker_line_width=1.5, opacity=0.6, texttemplate='%{text:.2s}',
                  textposition='outside')
fig_analise_saldo.update_yaxes(tickformat='R$,.2f')
fig_analise_saldo.show()
```

```python
fig_analise_segmento = px.pie(df_analise_segmento, names='Segmento', values="var_acoes", title='Varaição R$ por Segmento')
fig_analise_segmento.show()
```

```python
df_analise_faixa_etaria = df_principal.groupby("Cat_idade")['var_acoes'].sum().reset_index()
df_analise_faixa_etaria
```

```python
fig_cat_idade = px.bar(df_analise_faixa_etaria, x='Cat_idade', y='var_acoes', text='var_acoes',
                       title="Variação R$ por Categoria de Idade")
fig_cat_idade.update_traces(marker_color='rgb(251,180,174)', marker_line_color='#fd3216',
                  marker_line_width=1.5, opacity=0.6)
fig_cat_idade.show()
```
